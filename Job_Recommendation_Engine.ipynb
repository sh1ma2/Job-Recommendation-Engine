```python
data1['Profile'] = data1['Job Experience Required'] + data1['Key Skills'] + data1['Role Category'] + data1['Functional Area'] + data1['Industry']
data1.head()
```

```python
new = data1.drop(columns=['Job Experience Required','Key Skills','Role Category','Functional Area','Industry'])
new.head()
```

```python
new = new.rename({'Job Title': 'Job_Title'}, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=500,stop_words='english')
```

```python
vector = cv.fit_transform(new['Profile']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(Job):
    index = new[new['Job_Title'] == Job].index[0]
    distances = similarity[index]
    jobs_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in jobs_list:
        print(new.iloc[i[0]].Job_Title)
```

```python
recommend('Data Analyst')
```

```python
recommend('Testing Engineer')
```

```python
recommend('Sales Executive/Officer')
```
